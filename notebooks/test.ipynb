```python
import pandas as pd

def add_ordered_dict_column(df, cols, new_col_name):
    """
    Ajoute une colonne contenant des dictionnaires ordonnés des valeurs 
    des colonnes spécifiées, triées de la plus grande à la plus petite.

    Parameters:
    - df (pd.DataFrame): Le DataFrame d'origine.
    - cols (list): Liste des colonnes à inclure dans le dictionnaire.
    - new_col_name (str): Nom de la nouvelle colonne à ajouter.

    Returns:
    - pd.DataFrame: Le DataFrame avec la nouvelle colonne ajoutée.
    """
    def create_ordered_dict(row):
        # Crée un dictionnaire trié en fonction des valeurs
        return dict(sorted(((col, row[col]) for col in cols), key=lambda x: x[1], reverse=True))

    # Ajoute la nouvelle colonne
    df[new_col_name] = df.apply(create_ordered_dict, axis=1)
    return df

# Exemple d'utilisation
data = {
    'Value2pts': [5, 12, 8],
    'Value3pts': [7, 10, 9],
    'ValueLF': [3, 15, 6]
}
df = pd.DataFrame(data)

# Ajout de la nouvelle colonne avec la fonction
cols_to_order = ['Value2pts', 'Value3pts', 'ValueLF']
df = add_ordered_dict_column(df_shoot, cols_to_order, 'OrderedValues')

print(df)
```

```python
def calculate_percentages(df, new_col_name, cols):
    """
    Calcule le pourcentage d'individus pour lesquels chaque colonne est en première position.

    Parameters:
    - df (pd.DataFrame): Le DataFrame contenant la colonne des dictionnaires ordonnés.
    - new_col_name (str): Nom de la colonne contenant les dictionnaires ordonnés.
    - cols (list): Liste des colonnes à vérifier.

    Returns:
    - dict: Pourcentage pour chaque colonne.
    """
    first_column_counts = {col: 0 for col in cols}

    for ordered_dict in df[new_col_name]:
        first_col = next(iter(ordered_dict))  # Récupère la première clé du dictionnaire
        if first_col in first_column_counts:
            first_column_counts[first_col] += 1

    total = len(df)
    percentages = {col: (count / total) * 100 for col, count in first_column_counts.items()}
    return percentages
```

```python
import os
import pandas as pd

def process_and_save_individual_files(folder_path, output_folder):
    """
    Charge chaque fichier CSV dans un dossier, ajoute une colonne 'année' basée sur le nom du fichier,
    et sauvegarde le fichier modifié.

    Parameters:
    - folder_path (str): Chemin vers le dossier contenant les fichiers CSV.
    - output_folder (str): Chemin vers le dossier où enregistrer les fichiers modifiés.

    Returns:
    - None
    """
    # Créer le dossier de destination s'il n'existe pas
    os.makedirs(output_folder, exist_ok=True)

    # Parcourir tous les fichiers dans le dossier
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):  # Vérifie que le fichier est un CSV
            file_path = os.path.join(folder_path, filename)
            
            # Lire le fichier CSV
            df = pd.read_csv(file_path)
            
            # Extraire l'année du nom de fichier (e.g., "stat_per_game_1980.csv" -> 1980)
            year = int(filename.split('_')[-1].split('.')[0])
            
            # Ajouter la colonne 'année'
            df['Year'] = year
            
            # Définir le chemin de sortie pour le fichier modifié
            output_file_path = os.path.join(output_folder, filename)
            
            # Sauvegarder le fichier modifié
            df.to_csv(output_file_path, index=False)
            print(f"Fichier traité et sauvegardé : {output_file_path}")

# Exemple d'utilisation
folder_path = r"C:\Users\carlf\Documents\GitHub\The_Great_MVP_Predictor\stats_csv"
output_folder = r"C:\Users\carlf\Documents\GitHub\The_Great_MVP_Predictor\stats_csv"

# Appeler la fonction
process_and_save_individual_files(folder_path, output_folder)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep

def scrape_basketball_contracts():
    # URL cible
    url_start = "https://www.basketball-reference.com/contracts/players.html"
    
    # En-têtes pour simuler un navigateur web
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    try:
        # Faire la requête HTTP
        sleep(2)  # Délai de politesse
        response = requests.get(url_start, headers=headers)
        response.raise_for_status()  # Vérifier si la requête a réussi
        
        # Parser le HTML
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Trouver la table des contrats (id="player-contracts")
        contracts_table = soup.find('table', {'id': 'player-contracts'})
        
        if contracts_table is None:
            raise ValueError("Table des contrats non trouvée sur la page")
            
        # Extraire les en-têtes
        headers = []
        for th in contracts_table.find_all('th'):
            headers.append(th.text.strip())
            
        # Extraire les données
        rows = []
        for tr in contracts_table.find_all('tr')[1:]:  # Ignorer la première ligne (en-têtes)
            row = []
            for td in tr.find_all(['td', 'th']):
                row.append(td.text.strip())
            if row:  # Ignorer les lignes vides
                rows.append(row)
                
        # Créer un DataFrame
        df = pd.DataFrame(rows, columns=headers)
        
        # # Nettoyer les données (supprimer les caractères spéciaux des montants)
        # money_columns = df.columns[df.columns.str.contains('20')]  # Colonnes des années
        # for col in money_columns:
        #     df[col] = df[col].str.replace('$', '').str.replace(',', '').astype(float)
            
        return df
        
    except requests.exceptions.RequestException as e:
        print(f"Erreur lors de la requête HTTP: {e}")
        return None
    except Exception as e:
        print(f"Une erreur s'est produite: {e}")
        return None
df = scrape_basketball_contracts()
```

```python
from nba_api.stats.static import players
from nba_api.stats.endpoints import commonplayerinfo
import json

# Fonction pour afficher les informations détaillées sur un joueur
def get_player_info(player_name):
    # Rechercher le joueur par nom
    player_dict = players.get_players()
    player = [p for p in player_dict if p['full_name'].lower() == player_name.lower()]
    
    if not player:
        print(f"No player found with the name: {player_name}")
        return

    player_id = player[0]['id']  # ID du joueur
    print(f"Player found: {player[0]['full_name']} (ID: {player_id})")
    
    # Récupérer les informations détaillées sur le joueur
    player_info = commonplayerinfo.CommonPlayerInfo(player_id=player_id)
    data = player_info.get_normalized_dict()  # Normaliser les données
    
    # Afficher les données sous format JSON pour une meilleure lisibilité
    print(json.dumps(data, indent=4))

# Exemple d'utilisation
player_name = "LeBron James"  # Nom du joueur
get_player_info(player_name)
```

```python
import os
import pandas as pd

def remove_league_average_rows(folder_path):
    """
    Parcourt tous les fichiers CSV dans un dossier, supprime les lignes où la colonne 'Player' est égale à 'League Average',
    et réenregistre les fichiers nettoyés.

    Parameters:
    - folder_path (str): Chemin du dossier contenant les fichiers CSV.

    Returns:
    - None
    """
    # Parcourir tous les fichiers du dossier
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):  # Vérifie que le fichier est un CSV
            file_path = os.path.join(folder_path, filename)
            
            # Lire le fichier CSV
            try:
                df = pd.read_csv(file_path)
                
                # Vérifier si la colonne 'Player' existe
                if 'Player' in df.columns:
                    # Supprimer les lignes où 'Player' == 'League Average'
                    df = df[df['Player'] != 'League Average']
                    
                    # Réenregistrer le fichier nettoyé
                    df.to_csv(file_path, index=False)
                    print(f"Fichier traité et sauvegardé : {file_path}")
                else:
                    print(f"La colonne 'rk' est absente dans : {file_path}")
            except Exception as e:
                print(f"Erreur lors du traitement du fichier {file_path}: {e}")

# Exemple d'utilisation
folder_path = r"C:\Users\carlf\Documents\GitHub\The_Great_MVP_Predictor\data\stats_csv"
remove_league_average_rows(folder_path)
```

```python
import requests
import csv
from nba_api.stats.static import players
from nba_api.stats.endpoints import commonplayerinfo

# Fetch NBA player metadata
nba_players = players.get_players()

# Divide the list into three parts
total_players = len(nba_players)
chunk_size = total_players // 3
nba_players_parts = [
    nba_players[:chunk_size],  # First part
    nba_players[chunk_size:2 * chunk_size],  # Second part
    nba_players[2 * chunk_size:]  # Third part
]
```

```python
# Base URL for player images
base_image_url = "https://cdn.nba.com/headshots/nba/latest/1040x760/{player_id}.png"

# Function to scrape player data for a given chunk
def scrape_player_data(player_chunk, output_file):
    player_data = []
    for player in player_chunk:
        player_id = player['id']
        image_url = base_image_url.format(player_id=player_id)
        # Fetch additional temporal information
        try:
            player_info = commonplayerinfo.CommonPlayerInfo(player_id=player_id).get_dict()
            player_stats = player_info['resultSets'][0]['rowSet'][0]
            
            # Extract temporal information
            from_year = player_stats[22]  # FROM_YEAR
            to_year = player_stats[23]    # TO_YEAR
            draft_year = player_stats[27]  # DRAFT_YEAR
        except Exception as e:
            from_year = None
            to_year = None
            draft_year = None
            print(f"Error fetching data for player {player['full_name']}: {e}")
        
        # Check if the image URL is valid
        response = requests.head(image_url)
        if response.status_code == 200:
            player_data.append({
                "Player Name": player['full_name'],
                "Player ID": player_id,
                "Image URL": image_url,
                "From Year": from_year,
                "To Year": to_year,
                "Draft Year": draft_year,
            })
    
    # Save the chunk data to a CSV
    csv_columns = ["Player Name", "Player ID", "Image URL", "From Year", "To Year", "Draft Year"]
    try:
        with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
            writer.writeheader()
            writer.writerows(player_data)
        print(f"CSV file '{output_file}' created successfully!")
    except Exception as e:
        print(f"Error while creating CSV: {e}")

# Scrape data for each part and save to separate CSV files
scrape_player_data(nba_players_parts[0], "nba_players_part1.csv")
# scrape_player_data(nba_players_parts[1], "nba_players_part2.csv")
# scrape_player_data(nba_players_parts[2], "nba_players_part3.csv")
```
